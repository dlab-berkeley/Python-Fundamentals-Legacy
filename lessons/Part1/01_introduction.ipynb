## What it means to "know how to program"

Most programmers can program in more than one language. That's because they know *how to program* generally, as opposed to "knowing" Python, R, Matlab, or any other language.

In other words, programming is an extendible skill. Basic programming concepts - conditionals, for loops, functions - can be found in almost any programming language with small variations. Thus "knowing how to program" is not an exercise in memorization. 

Even within Python, there are too many packages and functions to memorize. Rather, a programmer knows 1) general structures and programming logic, 2) how to find and use new functions, and 3) how to work through problems that arise. 

Thus, we **don't learn specific programming languages; we learn how to program.** That being said, Python is a general-purpose, powerful, and high-level programming language, which makes it a very useful language to be comfortable with.

## Coding Strategies

Planning ahead can help mitigate time spent dealing with bugs and errors in code. General steps for defensive coding are:

1. State the goals of your code as clearly as possible.
2. Plan out the general logic of steps needed to achieve the goal.
3. Translate the steps into code:
    a. Build up steps piece by piece.
    b. Test frequently to make sure code is working as expected and handle bugs as quickly as possible.

For each option, it is useful to predict what you think the output should look like, then compare that to the output of your code. These steps can help you code more effectively and make it easier to deal with issues that do come up, but there is still no way to entirely avoid bugs.

## Debugging

Here's a useful mental workflow to keep in mind when you want to try and debug an error:

1. Read the errors!
2. Read the documentation.
3. Make it smaller.
4. Figure out what changed.
5. Check your syntax.
6. Print statements!
